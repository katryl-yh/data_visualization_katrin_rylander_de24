## Exploratory DA 

```python
import pandas as pd

df = pd.read_csv("data/influensa.csv",delimiter=";")
df.head()
```

```python
df_antal = df[df["Mått"]== "Antal"]
df_antal.head()
```

```python
df_antal_sorterad = df_antal.sort_values(["Vårdform", "Datum"])
df_antal_sorterad.head()
```

```python
import plotly.express as px

fig = px.line(
    df_antal_sorterad,
    x="Datum",
    y="DATA",
    color="Vårdform",
    #markers=True,
    #log_y=True,
)

fig.show()
```

```python
df_per_100k = df[df["Mått"]== "Per 100 000"]
df_per_100k_sorterad = df_per_100k.sort_values(["Vårdform", "Datum"])
df_per_100k_sorterad.head()
```
